# Oracle DB example with cx_Oracle


```python
import cx_Oracle
con = cx_Oracle.connect('hr/oracle@localhost:1522/ORCLPDB1')
cur = con.cursor()
cur.arraysize = 100
cur.execute('select * from employees order by department_id')
res = cur.fetchall()
for r in res:
    print(r)
cur.close()
con.close()
```
